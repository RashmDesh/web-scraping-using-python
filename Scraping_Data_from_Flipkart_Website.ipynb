# **Scraping Data from Flipkart Website**


### Accessing the HTML content from webpage

* Here we will scrap the mobile phone prices, ratings, and model name 
from Flipkart, which is one of the popular e-commerce websites.
* I am going to inspect this flipkar page,by right-click on the element and click "inspect".
* After that Extracting the Name, Price and Rating, which are contained in the "div" tag, respectively.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

myurl = "https://www.flipkart.com/search?q=iphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=offs"
r = requests.get(myurl)

  
```

```python
#Parsing the HTML content
content = BeautifulSoup(r.content, 'html5lib')
print(content.prettify())
```

```python
name=content.find_all('div',{"class":"_4rR01T"})
price=content.find_all('div',{"class":"_30jeq3 _1_WHN1"})
rating=content.find_all('div',{"class":"_3LWZlK"})
```

```python
print(name[0].text)
print(price[0].text)
print(rating[0].text)
```

```python
nm=[]
pr=[]
rate=[]

for i in name:
    nm.append(i.text)
for i in price:
    pr.append(i.text)
for i in range(len(nm)):
    rate.append(rating[i].text)
```

```python
data={'Name':nm,'Price':pr,'Rating':rate}
df=pd.DataFrame(data)
print(df)
```
